```python
import pandas as pd
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt

# Load your CSV data into a DataFrame
df = pd.read_csv('kalibrasi 40 keempat.csv')
```

```python
# Assuming 'speed' is the column with your radar data
speed_data = df['Speed'].values
speed_data
```

```python
# Perform FFT
fft_result = fft(speed_data)

# Get the absolute value to obtain the magnitude
magnitude = np.abs(fft_result)

# Calculate the frequency bins
sample_spacing = 1000 # This depends on your data's sampling rate
freq = np.fft.fftfreq(speed_data.size, d=sample_spacing)
```

```python
# Find the index of the peak
peak_index = np.argmax(magnitude)

# Find the dominant frequency
dominant_frequency = freq[peak_index]

# Find the magnitude at the dominant frequency
dominant_magnitude = magnitude[peak_index]

print('Data Peak : ',peak_index)
print('frekuensi dominan :',dominant_frequency)
print('dominant magnitude:',dominant_magnitude)
```

```python
# Plot the frequency response
plt.figure(figsize=(12, 6))
plt.plot(freq, magnitude)
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
